```python
import shutil
import random

# Set the paths for your prediction directories
prediction_dir = 'Prediction/'
training_dir = 'Training/'

# Set the desired number of images in the test set
total_pred_images = 100

# Set the number of classes
num_classes = 4

# Calculate the target number of images per class for the test set
target_images_per_class = total_pred_images // num_classes

# Create the test directory if it doesn't exist
if not os.path.exists(prediction_dir):
    os.makedirs(prediction_dir)

# Iterate through the training directory and split the images
for class_name in os.listdir(training_dir):
    class_dir = os.path.join(training_dir, class_name)
    if os.path.isdir(class_dir):
        images = os.listdir(class_dir)
        random.shuffle(images)  # Shuffle the images randomly
        selected_images = images[:target_images_per_class]  # Select the desired number of images

        # Copy the selected images to the test set directory
        for image in selected_images:
            src = os.path.join(class_dir, image)
            dst = os.path.join(prediction_dir, class_name, image)
            os.makedirs(os.path.join(prediction_dir, class_name), exist_ok=True)
            shutil.copy(src, dst)

        # Delete the selected images from the training directory
        for image in selected_images:
            image_path = os.path.join(class_dir, image)
            os.remove(image_path)
```
